#  Exploring correlation between city population and review ratings

```python
Author: Caleb Sherry
```

```python
# Pulling the census data into a dataframe

census_df = pd.read_csv("CBDataFileExpanded.cs.csv")
census_df
```

```python
# Merging the census data with the reviews data to add population to our review data set

correlation_df = pd.merge(census_df, city_mean, on="City")
correlation_df.head(10)
```

```python
# Plotting the population vs average rating for each city to explore the data
# Noticed the outlier of having to use NYC population for Brooklyn

x_values = correlation_df["Pop2019"]
y_values = correlation_df["Rating"]
plt.scatter(x_values,y_values)
plt.xlabel('Population')
plt.ylabel('Avg Rating')
plt.title("City Population vs Avg. Review Ratings")
plt.show()
```

```python
# Dropping the outlier

correlation_df1 = correlation_df["Pop2019"] < 8000000
correlation_df1
correlation_df1=correlation_df[(correlation_df['Pop2019']<8000000)]
```

```python
# Plotting the new scatter plot, running the regression, and plotting the regression line and equation

x_values = correlation_df1["Pop2019"]
y_values = correlation_df1["Rating"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,10)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")

plt.annotate(line_eq,(10,8),xytext=(0,7),fontsize=15,color="red")
plt.xlabel('Population')
plt.ylabel('Avg Rating')
print(f"The r-squared is: {rvalue**2}")
plt.title("City Population vs Avg. Review Ratings")
plt.show()
```
