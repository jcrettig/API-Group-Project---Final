```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census
import gmaps

# Census & gmaps API Keys
from api_keys import (gkey)
c = Census(gkey, year=2013)

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame for Census Data
census_csv = "Heat Map2/CBDataFileExpanded.csv"
census_df = pd.read_csv(census_csv)
print(census_df)
```

```python
city_grid_csv = 'Heat Map2/rating_city_grid_original.csv'
city_grid_df = pd.read_csv(city_grid_csv)

final_city_df = city_grid_df.dropna()
final_city_df
```

```python
final_city_df = city_grid_df[city_grid_df['City'].isin(["Lockhart", "Memphis","Kansas City","Lexington", "Austin", "St Louis"
          , "Dallas", "Nashville","Atlanta", "Brooklyn"] )]
final_city_df
```

```python
grid_df = pd.DataFrame({
    
    'City Grid Rating': final_city_df, 
    
})
grid_df
```

```python
merge_df = pd.merge(grid_df, census_df, on='City')
merge_df
```

```python
#Create a list of Cities and States to pull coordinates from Google Place Search
City_df = pd.DataFrame(
    {"City":['Lockhart','Memphis', "Kansas City", 'Lexington','Austin', 'St Louis', 'Dallas', 'Nashville', 'Atlanta',
              'Brooklyn'],
     "State":['TX', 'TN', 'MO', 'NC', 'TX', 'MO', 'TX', 'TN', 'GA', 'NY']
    }
) 

City_df.head(10)
```

```python
base_url = "https://maps.googleapis.com/maps/api/geocode/json?"

Cities = ['{}, {}'.format(City, State) for City, State in zip(City_df['City'], City_df['State'])]
p_requests = [requests.get(base_url, params={
    'address':City,
    'key':gkey
}) for City in Cities]


places_data = [place.json() for place in p_requests]
places_data
```

```python
[place['results'][0] for place in places_data if len(place['results']) > 0]

#city_list = [h_attribute['results'][0]['address_components']['long_name'] for h_attribute in places_data if len(h_attribute['results']) > 0]
lat_list = [h_attribute['results'][0]['geometry']['location']['lat'] for h_attribute in places_data if len(h_attribute['results']) > 0]
lng_list = [h_attribute['results'][0]['geometry']['location']['lng'] for h_attribute in places_data if len(h_attribute['results']) > 0]

coordinates_df = pd.DataFrame(data={'Lat': lat_list, 'Lng': lng_list }, copy=True)
coordinates_df.head(10)
```

```python
# Merge the SFact_DF and line_DF into a combined dataframe
coordinates_df = pd.merge(City_df, coordinates_df, left_index = True, right_index = True)
coordinates_df.head(11)
```

```python
merge_df2 = merge_df.copy()
merge_df2[['City','State', 'Lat', 'Lng']]
```

```python
for i,r in coordinates_df.iterrows():
     merge_df2.loc[merge_df2['City']==r['City'],'Lat'] = r['Lat']
     merge_df2.loc[merge_df2['City']==r['City'],'Lng'] = r['Lng']
     merge_df2.loc[merge_df2['City']==r['City'],'State'] = r['State']
 
    
```

```python
locations = merge_df2[['Lat', 'Lng']]
locations
```

```python
markers = gmaps.marker_layer(locations)

# Adding city marks to the heatmap
info_box_template = """
<dl>
<dt>City</dt><dd>{City}</dd>
<dt>City Grid Rating</dt><dd>{City Grid Rating}</dd>
<dt>Population Count</dt><dd>{Pop2019}</dd>
<dt>Food Sales</dt><dd>{A&FSales1K}</dd>
<dt>Per Capita Income</dt><dd>{PerCapInc}</dd>
</dl>
"""

merge_info = [info_box_template.format(**row) for index, row in merge_df2.iterrows()]


# Create city symbol layer
merge_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=merge_info
)

```

```python
fig=gmaps.figure()
# Add marker layer and info box content ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content = merge_info)

# Add the layer to the map
fig.add_layer(markers)
             

# Display Map
fig
```
